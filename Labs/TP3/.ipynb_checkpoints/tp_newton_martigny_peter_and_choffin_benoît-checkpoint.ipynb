# Part 0: Demo using Gradient descent

First import the necessary libraries:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
import time
%matplotlib inline
```

Now import the necessary function from the optim_utils.py file.

```python
from optim_utils import test_solver
```

You'll have work only with the `test_solver` function.

This function expects a function as parameter.

The signature of the function `optimizer` to pass should be the following:

`optimizer(x0, f, f_grad, f_hessian)`

Let us now make an example with a gradient descent.

```python
def gradient_descent(x0, f, f_grad, f_hessian=None):

    default_step = 0.01
    c1 = 0.0001
    c2 = 0.9
    max_iter = 100
    
    # This variable is used to indicate whether or not we want to print
    # monitoring information (iteration counter, function value and norm of the gradient)
    verbose = False

    all_x_k, all_f_k = list(), list()
    x = x0

    all_x_k.append(x.copy())
    all_f_k.append(f(x))

    for k in range(1, max_iter + 1):

        grad_x = f_grad(x)

        # Compute a step size using a line_search to satisfy the
        # strong Wolfe conditions
        step, _, _, new_f, _, new_grad = optimize.line_search(f, f_grad, x,
                                                              -grad_x, grad_x,
                                                              c1=c1, c2=c2)
        if step is None:
            print("Line search did not converge at iteration %s" % k)
            step = default_step

        x -= step * grad_x

        all_x_k.append(x.copy())
        all_f_k.append(new_f)

        l_inf_norm_grad = np.max(np.abs(new_grad))

        if verbose:
            print('iter: %d, f: %.6g, l_inf_norm(grad): %.6g' %
                  (k, new_f, l_inf_norm_grad))

        if l_inf_norm_grad < 1e-6:
            break

    return np.array(all_x_k), np.array(all_f_k)
```

Now I can call the `test_solver` function with this solver:

```python
start = time.time()
test_solver(gradient_descent)
print('The running time of the three solvers is: ', time.time() - start)
```

# Part 2: Implement Newton method

#### Newton 1: linalg.solve

```python
import scipy.linalg as linalg
```

```python
def newton_1(x0, f, f_grad, f_hessian):
    default_step = 0.01
    c1 = 0.0001
    c2 = 0.9
    max_iter = 100
    
    # This variable is used to indicate whether or not we want to print
    # monitoring information (iteration counter, function value and norm of the gradient)
    verbose = True

    all_x_k, all_f_k = list(), list()
    x = x0

    all_x_k.append(x.copy())
    all_f_k.append(f(x))

    for k in range(1, max_iter + 1):

        grad_x = f_grad(x)
        
        ####################################
        # Compute here the search direction
        d = -linalg.solve(f_hessian(x), grad_x)
        ####################################

        # Compute a step size using a line_search to satisfy the
        # strong Wolfe conditions
        step, _, _, new_f, _, new_grad = optimize.line_search(f, f_grad, x,
                                                              d, grad_x,
                                                              c1=c1, c2=c2)
        if step is None:
            print("Line search did not converge at iteration %s" % k)
            step = default_step

        ##################################
        # Compute here the new value of x
        x = x + step*d
        ##################################

        all_x_k.append(x.copy())
        all_f_k.append(new_f)

        l_inf_norm_grad = np.max(np.abs(new_grad))

        if verbose:
            print('iter: %d, f: %.6g, l_inf_norm(grad): %.6g' %
                  (k, new_f, l_inf_norm_grad))

        if l_inf_norm_grad < 1e-6:
            break

    return np.array(all_x_k), np.array(all_f_k)
```

```python
start = time.time()
test_solver(newton_1)
print('The running time of the three solvers is: ', time.time() - start)
```

#### Nexton 2 (sparse.linalg)

```python
import scipy.sparse.linalg as sparse
```

```python
def newton_2(x0, f, f_grad, f_hessian):
    default_step = 0.01
    c1 = 0.0001
    c2 = 0.9
    max_iter = 100
    
    # This variable is used to indicate whether or not we want to print
    # monitoring information (iteration counter, function value and norm of the gradient)
    verbose = True

    all_x_k, all_f_k = list(), list()
    x = x0

    all_x_k.append(x.copy())
    all_f_k.append(f(x))

    for k in range(1, max_iter + 1):

        grad_x = f_grad(x)
        
        ####################################
        # Compute here the search direction
        if k > 1:
            d = -sparse.cg(f_hessian(x), grad_x, d)[0]
        else:
            d = -sparse.cg(f_hessian(x), grad_x)[0]
        ####################################

        # Compute a step size using a line_search to satisfy the
        # strong Wolfe conditions
        step, _, _, new_f, _, new_grad = optimize.line_search(f, f_grad, x,
                                                              d, grad_x,
                                                              c1=c1, c2=c2)
        if step is None:
            print("Line search did not converge at iteration %s" % k)
            step = default_step

        ##################################
        # Compute here the new value of x
        x = x + step*d
        ##################################

        all_x_k.append(x.copy())
        all_f_k.append(new_f)

        l_inf_norm_grad = np.max(np.abs(new_grad))

        if verbose:
            print('iter: %d, f: %.6g, l_inf_norm(grad): %.6g' %
                  (k, new_f, l_inf_norm_grad))

        if l_inf_norm_grad < 1e-6:
            break

    return np.array(all_x_k), np.array(all_f_k)
```

```python
start = time.time()
test_solver(newton_2)
print('The running time of the three solvers is: ', time.time() - start)
```

We observe that, for both linear system solvers (linalg.solve and sparse.linalg.cg), the Newton algorithm converges in 1 iteration for the two first functions, which are quadratic. This is the consequence of slide 7 in the quasi-newton lecture. In this situation, Newton is not affected by the conditioning of the problem, and converges therefore way faster than gradient descent (cf above cell). 
For the last function, the Newton algorithm converges in 12 iterations. The number of iterations is not 1 because the function is not quadratic. However, the number of iterations remains small, because of the line search we performed in the loop.

# Part 2: Implement DFP algorithm

You now need to implement the DFP algorithm.

```python
def dfp(x0, f, f_grad, f_hessian):
    default_step = 0.01
    c1 = 0.0001
    c2 = 0.95
    max_iter = 200
    
    # This variable is used to indicate whether or not we want to print
    # monitoring information (iteration counter, function value and norm of the gradient)
    verbose = True

    all_x_k, all_f_k = list(), list()
    x = x0

    all_x_k.append(x.copy())
    all_f_k.append(f(x))

    B = np.eye(len(x))  # inverse Hessian approximation
    
    grad_x = f_grad(x)
    
    for k in range(1, max_iter + 1):       
        
        ####################################
        # Compute here the search direction
        d = -np.dot(B, grad_x)
        ####################################

        # Compute a step size using a line_search to satisfy the
        # strong Wolfe conditions
        step, _, _, new_f, _, new_grad = optimize.line_search(f, f_grad, x,
                                                              d, grad_x,
                                                              c1=c1, c2=c2)
        
        if step is None:
            print("Line search did not converge at iteration %s" % k)
            step = default_step

        ##################################
        # Compute here the new value of x
        s = step*d
        x += s
        ##################################
        
        #########################################################
        # Update the inverse Hessian approximation
        y = new_grad - grad_x
        aux1 = (np.dot(s.reshape(len(s), 1), s.reshape(1, len(s))))/(np.dot(s, y))
        aux2 = (np.dot(B, np.dot(y.reshape(len(y), 1), np.dot(y.reshape(1, len(y)), B)))) / (np.dot(y, np.dot(B, y)))
        B += aux1 - aux2
        #########################################################
        
        all_x_k.append(x.copy())
        all_f_k.append(new_f)

        l_inf_norm_grad = np.max(np.abs(new_grad))

        if verbose:
            print('iter: %d, f: %.6g, l_inf_norm(grad): %.6g' %
                  (k, new_f, l_inf_norm_grad))

        if l_inf_norm_grad < 1e-6:
            break
            
        grad_x = new_grad

    return np.array(all_x_k), np.array(all_f_k)
```

```python
start = time.time()
test_solver(dfp)
print('The running time of the three solvers is: ', time.time() - start)
```

We observe that, for both quadratic functions, the algorithm converges in $6-8$ iterations. For the third function, the convergence speed is smaller, the algorithm converges in around 90 iterations.
We saw during the lecture that the DFP algorithm for quadratic functions is supposed to converge in $n$ iterations, where $n$ is the dimension of the $x$. Here, $n = 2$, so the algorithm is expected to converge in $2$ iterations for the $2$ quadratic functions, which is not the case. The reason is that we use a line search to have update of the step size $\rho_k$, which is an estimation of the optimal step size used in the theorem.

# Part 3: Implement BFGS algorithm

You now need to implement the BFGS algorithm.

```python
def bfgs(x0, f, f_grad, f_hessian):
    default_step = 0.01
    c1 = 0.0001
    c2 = 0.9
    max_iter = 100
    
    # This variable is used to indicate whether or not we want to print
    # monitoring information (iteration counter, function value and norm of the gradient)
    verbose = True

    all_x_k, all_f_k = list(), list()
    x = x0

    all_x_k.append(x.copy())
    all_f_k.append(f(x))

    H = np.eye(len(x))  # Hessian approximation
    
    grad_x = f_grad(x)
    
    for k in range(1, max_iter + 1):       
        
        ####################################
        # Compute here the search direction
        d = -linalg.solve(H, grad_x)
        ####################################

        # Compute a step size using a line_search to satisfy the
        # strong Wolfe conditions
        step, _, _, new_f, _, new_grad = optimize.line_search(f, f_grad, x,
                                                              d, grad_x,
                                                              c1=c1, c2=c2)
                
        if step is None:
            print("Line search did not converge at iteration %s" % k)
            step = default_step

        ##################################    
        # Compute here the new value of x
        s = step*d
        x += s
        ##################################
        
        #####################################################
        # Update the Hessian approximation   
        y = new_grad - grad_x
        aux1 = (np.dot(y.reshape(len(y), 1), y.reshape(1, len(y))))/(np.dot(y, s))
        aux2 = (np.dot(H, np.dot(s.reshape(len(s), 1), np.dot(s.reshape(1, len(s)), H)))) / (np.dot(s, np.dot(H, s)))
        H += aux1 - aux2
        #####################################################
        
        all_x_k.append(x.copy())
        all_f_k.append(new_f)

        l_inf_norm_grad = np.max(np.abs(new_grad))

        if verbose:
            print('iter: %d, f: %.6g, l_inf_norm(grad): %.6g' %
                  (k, new_f, l_inf_norm_grad))

        if l_inf_norm_grad < 1e-6:
            break
            
        grad_x = new_grad

    return np.array(all_x_k), np.array(all_f_k)
```

```python
start = time.time()
test_solver(bfgs)
print('The running time of the three solvers is: ', time.time() - start)
```

As for the DFP algorithm, we observe that, for the two quadratic functions, the number of iterations before convergence is very small, but not less than $n = 2$. However, BFGS converges faster for all functions to minimize than DFP, because it is less sensitive than DFP to the use of line search for approximate step size.

We define a test function for the scipy implementations of BFGS and L-BFGS

```python
def test_scipy_bfgs():

    x_min, x_max = -1, 2
    y_min, y_max = 2.25 / 3 * x_min - .2, 2.25 / 3 * x_max - .2
    x_min *= 1.2
    x_max *= 1.2
    y_min *= 1.2
    y_max *= 1.2
    fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 4))

    for index, [ax, (f, f_prime, f_hessian)] in enumerate(zip(axes, (
            mk_quad(.7),
            mk_quad(.02),
            (rosenbrock, rosenbrock_prime, rosenbrock_hessian)))):
        
        print("\nRunning solver on case %d" % (index + 1))
        
        logging_f = optim_utils.LoggingFunction(f)
        
        x0 = np.array([1.6, 1.1])
        
        all_x_k, all_f_k = fmin_bfgs(f = logging_f, x0 = x0, fprime=f_prime)
        all_x = np.array(logging_f.all_x)
        
        optim_utils.plot_convergence(f, ax, all_x, all_f_k, all_x, x_min, x_max, y_min, y_max)
   
    plt.show()
```
