```python
%matplotlib inline
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy import linalg
```

```python
from sklearn import datasets
from sklearn.linear_model import ridge_regression, LogisticRegression

iris = datasets.load_iris()
X = iris.data
y = iris.target

# Take only 2 features
X = X[:, :2]

# And add a column of 1 for the intercept
X = np.concatenate((X, np.ones((X.shape[0], 1))), axis=1)

# Make it binary
X = X[y < 2]
y = y[y < 2]

y[y == 0] = -1
```

```python
plt.scatter(X[y > 0, 0], X[y > 0, 1], color='r')
plt.scatter(X[y < 0, 0], X[y < 0, 1], color='g')
```

## Now implement Newton method for logistic regression

```python
def newton_logistic(X, y, alpha=1.):
    X = np.asarray(X, dtype=np.float)
    y = np.asarray(y, dtype=np.float)

    n_samples, n_features = X.shape

    w = np.zeros(n_features)
    pobj = []

    for k in range(20):
        ywTx = y * np.dot(X, w)
        temp = 1. / (1. + np.exp(ywTx))
        grad = - np.dot(X.T, (y * temp)) + alpha * w
        hess = np.dot(X.T, (temp * ( 1. - temp ))[:, None] * X)
        hess.flat[::n_features + 1] += alpha

        hess[-1, -1] -= alpha  # don't penalize intercept
        grad[-1] -= alpha * w[-1]

        w -= linalg.solve(hess, grad)

        this_pobj = np.sum(np.log( 1. + np.exp( - y * np.dot(X, w))))
        this_pobj += alpha * np.dot(w[:-1], w[:-1]) / 2.
        pobj.append(this_pobj)

    print("Global minimum : %s" % pobj[-1])

    return w, pobj
```

```python
alpha = 0.1
w, pobj = newton_logistic(X, y, alpha=alpha)
```

### did cost function go down?

```python
#plt.plot(np.log10(pobj[:-1] - pobj[-1]), 'r')
plt.plot(np.log10(pobj), 'r')
plt.xlabel('Iterations')
plt.ylabel('Primal')
plt.show()
```

### Does it do the job?

```python
plt.scatter(X[y > 0, 0], X[y > 0, 1], color='r')
plt.scatter(X[y < 0, 0], X[y < 0, 1], color='g')
xx = np.linspace(4, 8, 10)
plt.plot(xx,  - xx * w[0] / w[1] - w[2] / w[1], 'k')
```

# Logistic Regression with BFGS and L-BFGS

```python
all_pobj = []

def f(w):
    pobj = np.sum(np.log(1. + np.exp(- y * np.dot(X, w))))
    pobj += alpha * np.dot(w[:-1], w[:-1]) / 2.
    all_pobj.append(pobj)
    return pobj

def fprime(w):
    ywTx = y * np.dot(X, w)
    temp = 1. / (1. + np.exp(ywTx))
    grad = -np.dot(X.T, (y * temp)) + alpha * w
    grad[-1] -= alpha * w[-1]  # take care of intercept
    return grad

from scipy.optimize import fmin_bfgs, fmin_l_bfgs_b
w = fmin_bfgs(f, x0=np.zeros(X.shape[1]), fprime=fprime)
```

```python
plt.plot(np.log10(all_pobj), 'r')
plt.xlabel('Iterations')
plt.ylabel('Primal')
plt.show()
```

```python
plt.scatter(X[y > 0, 0], X[y > 0, 1], color='r')
plt.scatter(X[y < 0, 0], X[y < 0, 1], color='g')
xx = np.linspace(4, 8, 10)
plt.plot(xx,  - xx * w[0] / w[1] - w[2] / w[1], 'k')
```

### now with L-BFGS-B

```python
from scipy.optimize import fmin_l_bfgs_b
w, _, _ = fmin_l_bfgs_b(f, x0=np.zeros(X.shape[1]), fprime=fprime)
```
