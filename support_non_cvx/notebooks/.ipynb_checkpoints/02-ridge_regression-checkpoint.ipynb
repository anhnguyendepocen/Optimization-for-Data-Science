```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import linalg
```

## Generate simulated data

$y = xw+b+noise$

```python
np.random.seed(0)
n_samples, n_features = 100, 1
X = np.random.randn(n_samples, n_features)
w = np.random.randn(n_features)
b = 10.
y = np.dot(X, w) + b
y += 0.3 * np.random.randn(n_samples)
```

```python
def plot_data(w, b=0.):
    plt.plot(X[:, 0], y, 'o', alpha=0.2)

    xx = np.linspace(-2, 2, 100)
    yy = np.dot(xx[:, np.newaxis], w) + b
    plt.plot(xx, yy, 'k')
    plt.grid('on')

plot_data(w, b)
```

## Option 1 : center X and y

$\min_w \|y - Xw - b\|$

solved with `linalg.solve`

```python
X_mean = np.mean(X, axis=0)
y_mean = np.mean(y)

X_centered = X - X_mean
y_centered = y - y_mean

w_hat = linalg.solve(np.dot(X_centered.T, X_centered), np.dot(X_centered.T, y))
b_hat = - np.dot(X_mean, w_hat) + y_mean

plot_data(w_hat, b_hat)
```

## Option 2: add a column with ones

```python
Xb = np.concatenate((X, np.ones((n_samples, 1))), axis=1)

# and solve in dimension p + 1
w_hat, b_hat = linalg.solve(np.dot(Xb.T, Xb), np.dot(Xb.T, y))
plot_data(w_hat, b_hat)
```
